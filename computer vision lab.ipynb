# 1. Ethical problems with image recognition

The Ethical Questions That Haunt Facial-recognition Research - Nature.com, 2020

https://www.nature.com/articles/d41586-020-03187-3

- surveillance cameras were used to employ machine learning algorithms recognizing personal characteristics
- the dataset was trained on a large pool of Uyghur people photographs
- the article detailing the process eventually got published in a scientific journal
- loud criticism from scientist community ensued
- the dataset was collected without consent of individuals, and this problematic practice has happened multiple times before
- as a well known learning problemn, reinforcing datasets can lead to automatized biases in detection
- seems to be a divisive topic among researchers on how much of an actual ethical problems could stem from this practice
- as a broader illustration of this problem, many freely available online photos can be easily used to train models
- after some press coverage eventually several university and Microsoft websites removed the datasets
- in most of Europe, there's a great deal of legal uncertainty around employing these practices as of 2020
- in the last few years lawsuits have resulted in financial compensation in the range of hundreds of millions of dollars
- a similar problem of consent with ethnical groups has resulted a retraction of DNA sequencing study of Uyghur individuals, apparently backlash from scientific, philosophical and ethical researchers and intellectuals can make a difference
- a closely related practice of facial recognition in crime prediction provided 80% accuracy in published studies
- consequently the application of algorithms can be useful in crime prevention and reduction
- however, the majority of respondents in a survey were about equally likely to say they supported informed consent in participation of data collection and freely used licensed data (40% and 36%), these numbers varied depending on the level of invasiveness and harm reduction weighing violation of personal space and human right problems
